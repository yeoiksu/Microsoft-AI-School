## 2. Web Crawling

```python
import bs4
import requests
```

```python
url ='https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

### BeautifulSoup(text, parser)

#### parser : 구조화 되있는 데이터를 이해하는 해석기

```python
parser = 'html.parser'
bsObj = bs4.BeautifulSoup(result, parser)
```

#### 원하는 정보 찾기

```python
# find : 첫번째 꺼를 찾음
# find('tag이름', {'class or key': 'class or key 이름'} )
bsObj.find('a', {'class': 'basicList_link__JLQJf'}).text
```

```python
# find_all : 해당조건을 만족하는 모든 요소를 리스트로 반환
items = bsObj.find_all('a', {'class': 'basicList_link__JLQJf'})
```

```python
# for item in items:
#     print(item)
#     print('-'*130)
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```
