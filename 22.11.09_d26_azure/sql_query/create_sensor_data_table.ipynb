```sql
CREATE TABLE dbo.SensorData(  --dbo: database owner
    id INT IDENTITY(1,1) NOT NULL,  --IDENTITY(n,m) n부터 m씩 증가
    DeviceID VARCHAR(20) NOT NULL,
    Temperature FLOAT NULL,  -- 온도
    Humidity FLOAT NULL,     -- 습도
    Dust FLOAT NULL,         -- 미세먼지 
    RegiDate DATETIME DEFAULT(GETDATE()),  -- Data의 시간 기록
    PRIMARY KEY CLUSTERED (id)  -- 무조건 not null & index 자동생성
                                -- CLUSTERED : 논리적으로 정렬 -> 속도 up
)
```

```sql
SELECT *
FROM dbo.SensorData
```
