## 3. Dimensionality Reduction

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# digit 관련 data 불러오기
digits = load_digits()
```

```python
print(digits.DESCR)
```

```python
data  = digits.data    # 8 X 8 data
label = digits.target  # 0에서 9의 숫자
```

```python
data.shape
```

```python
label.shape
```

64개의 data를 8 x 8 data를 바꾸고 싶을때는 reshape(8,8)함수사용

```python
data[0].reshape(8,8)
```

```python
label[0]
```

```python
plt.imshow(data[0].reshape(8,8))
print('Label: {}'.format(label[0]))
```

### PCA (주성분 분석)

```python
from sklearn.decomposition import PCA

# N차원으로 줄여줘
N = 2
pca = PCA(n_components= N)

new_data = pca.fit_transform(data)
```

```python
print('원본 데이터의 차원 \n{}\n'.format(data.shape))
print('PCA를 적용한 데이터의 차원 \n{}'.format(new_data.shape))
```

```python
# 2개
new_data[0]
```

```python
# 64개
data[0]
```

```python
# 데이터 분포
plt.scatter(new_data[:, 0],new_data[:, 1], c= label, alpha= 0.4)
plt.legend()
```
