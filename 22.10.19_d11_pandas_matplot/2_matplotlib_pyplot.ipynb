## 2. Matplotlib pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Style 적용하기
plt.style.use('classic')
```

```python
# Sample Data 생성
x = np.linspace(0, 10 ,100) # 0과 10사이에 100개의 데이터를 만들어라
```

```python
plt.figure(figsize=(3,2))
plt.plot(x,np.sin(x))
```

```python
plt.figure(figsize=(3,2))
plt.plot(x,np.cos(x))
```

#### 두 개의 그래프를 하나의 그래프로 그리기

```python
# 한 cell에 같이 넣음
fig = plt.figure(figsize=(4,3))
plt.plot(x,np.sin(x), '-')
plt.plot(x,np.cos(x), '--')
```

#### Figure 저장하기

```python
fig.savefig('sin_cos_figure.png')
```

#### subplot 이용하여 구역 따로 설정하기

```python
# (2,1)짜리에서 1번째
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

# (2,1)짜리에서 2번째
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
# plt.figure(figsize=(5,3))
plt.figure()
```

#### Grid line

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure(figsize=(5,3))
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))
```

#### 색 표현방법

```python
plt.figure(figsize=(5,3))
plt.plot(x, np.sin(x-0), color='red')          # 색상의 이름
plt.plot(x, np.sin(x-1), color='0.25')         # 0~1사이의 회색조 표현 
plt.plot(x, np.sin(x-2), color='g')            # 짧은 색상코드 RGB CMYK
plt.plot(x, np.sin(x-3), color='#FFDD44')      # 16진수로 표현한 RGB 
plt.plot(x, np.sin(x-4), color=(1,0, 0.5,0.3)) # RGB 값을 튜플로 표현
plt.plot(x, np.sin(x-5), color='chartreuse')   # 고유의 이름
```

#### 선 표현 방법

```python
# solid  : 실선          (-)
# dashed : 점선          (--)
# dashdot: 점 + 실선     (-.)
# dotted : 아주작은 점선 (:)
plt.figure(figsize=(5,4))
plt.plot(x, x+3, linestyle='dotted')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x-1, linestyle='-')
plt.plot(x, x-2, linestyle='--')
plt.plot(x, x-3, linestyle='-.')
plt.plot(x, x-4, linestyle=':')
```

#### 제일 많이 사용하는 방식

```python
plt.figure(figsize=(4,3))
plt.plot(x, x+3, ':k')  # 작은 점선 + black
plt.plot(x, x+2, '-.b') # 점실선    + blue 
plt.plot(x, x+1, '--c') # 점선      + cyan
plt.plot(x, x+0, '-g')  # 실선      + green
```

#### plot 조정하기 : 축 경계(xlim, ylim)

```python
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)       # x축 경계
plt.ylim(-1.2, 1.2)    # y축 경계
```

```python
# x축 뒤집기
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.xlim(11, -1)     # x축 뒤집어서
plt.ylim(-1.2, 1.2)     
```

```python
# y축 뒤집기
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)     
plt.ylim(1.2, -1.2) # y축 뒤집어서
```

```python
# 한번에 뒤집기 : 잘 사용안함
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.2,1.2])
```

```python
# tight : 꽉 차게 하게 나옴
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.axis('tight')    
```

```python
# equal : 현실에 가까운 그래프 확인
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.axis('equal') 
```

#### plot에 label 붙이기
    - title  : 그래프의 이름 
    - xlabel : x축 label
    - ylabel : y축 label

```python
plt.figure(figsize=(4,3))
plt.plot(x, np.sin(x))
plt.axis('equal') 

# 그래프 title
plt.title('A Sine Curve')

# x축 label
plt.xlabel('x')

# y축 label
plt.ylabel('sin(x)')
```

    - legend : 다양한 결과앖의 label을 지정할 때 유용 

```python
plt.figure(figsize=(6,4))
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '--b', label='cos(x)')
plt.axis('equal') 
plt.legend()
```

#### 산점도의 출력

```python
x = np.linspace(0,10,30)
y = np.sin(x)
```

```python
# marker : 표현하고 싶은 마커로 사용 ('o')
plt.figure(figsize=(4,3))
plt.plot(x, y, 'o', color='black')
plt.axis('equal') 
```

#### marker의 종류
    ['o', ',', 'x', '+', 'v', '<', '>', 's', 'd']

```python
# random 값을 만들어줌
rnd = np.random.RandomState(0)
markers = ['o', ',', 'x', '+', 'v', '<', '>', 's', 'd']

for marker in markers:
    plt.plot(rnd.rand(5), rnd.rand(5), marker, label= 'marker= {0}'.format(marker))
    plt.xlim(0, 1.4)
    plt.legend()
```

```python
# (선 종류, 산점도, 색깔)
plt.figure(figsize=(5,3))
plt.plot(x, y  , '-ok')   # 원형
plt.plot(x, y-1, '-p')    # 5각형
```

#### marker의 굵기, 색깔, 선의 굵기 조절

```python
plt.figure(figsize=(5,3))
plt.ylim(-1.2, 1.2)    
plt.plot(x, y, '-p', color='red', 
         markersize= 15,
         markerfacecolor= 'white',
         markeredgecolor= 'blue',
         markeredgewidth= 2,
         linewidth = 4
         )
```

#### scatter plot을 이용해서 산개도나 오차를 표현할 때 많이 사용

```python
plt.figure(figsize=(7,5))
plt.scatter(x, y, marker ='o',
            s = 150
           )
```
