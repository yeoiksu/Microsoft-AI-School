```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

## 1. 1차 함수 (y = ax+ b)

```python
def linear_function(x):
    a = 0.5   # 기울기
    b = 2     # y절편
    return a*x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

## 2. 2차 함수 (y = ax^2 + bx +c)

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    
    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2)) 
```

```python
y2 = quadratic_function(x)

plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

## 3. 3차 함수 (y = ax^3 + bx^2 + cx + d)

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a*x**3 + b*x**2 + c*x + d
```

```python
print(cubic_function(2)) 
```

```python
y3 = cubic_function(x)

plt.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

### 함수의 최소값 / 최대값

```python
# 2차 함수의 수식
def my_func(x):
    a = 1
    b = -3
    c = 10
    
    return a*x**2 + b*x + c 
```

```python
x = np.arange(-10, 10 ,0.1)
y = my_func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Function')
plt.scatter(1.5, my_func(1.5))
```

## 4. 지수 함수 (y = a^x)

```python
def exponential_function(x):
    a = 4
    
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3, 2 ,0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-4,3)
plt.ylim(-1,15)
plt.title('Exponential Function')
```
